```python
def plot_(df, xlabel, ylabel, title, path, size_index=1):
    import os, uuid
    plt.rcParams['axes.facecolor'] = 'white'
    sns.set_style("whitegrid")
    sns.set_palette("mako")
    plt.figure(figsize=(10*size_index,6*size_index))
    df_cp = df.copy().sort_index()
    years = df_cp.index
    rate = df_cp.values.reshape(-1)
    sns.barplot(x=years, y=rate, order=years, alpha=0.7, hue=rate, dodge=False)
    plt.title(title, fontsize=9*size_index)
    plt.xlabel(xlabel, fontsize=8*size_index)
    plt.ylabel(ylabel, fontsize=9*size_index)
    
    if os.path.exists(os.path.join(os.getcwd(),path)):
        path = os.path.join(os.getcwd(),path.format(uuid.uuid1()))
    
    else:
        path = os.path.join(os.getcwd(),path)
    
    plt.savefig(path)
    
    plt.show()
```

```python
plot_(bd_internet_penetration_rate,"Years","Internet Penetration Rate","Percantage of Population in Bangladesh who has access to Internet","Internet_Penetration_Rate_Bangladesh.jpg",size_index=2)
```

```python
plot_(bd_adult_literacy_rate, "Year", "Adult Literacy Rate %", "Adult Literacy Rate in Bangladesh", "adult_literacy.jpg", size_index=2)
```
